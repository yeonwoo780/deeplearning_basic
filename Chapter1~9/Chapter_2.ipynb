## 일차 함수
 - 일차 함수식 에서 a는 기울기, b는 절편이라고 함
 - x가 주어지고 원하는 y값이 있을 때 적절한 a와 b를 찾는 것(딥러닝의 간단한 설명)


## 이차함수
- 이차 함수란 y가 x에 관한 이차식으로 표현되는 경우를 말함
- a > 0이면 아래로 볼록한 그래프가 됨
- 포물선의 맨 아래에 위치한 지점이 최솟값이 되는데, 딥러닝을 실행할 때 이 최솟값을 찾아내는 과정이 매우 중요


- 딥러닝을 이해하는 데 가장 중요한 수학 원리는 **미분**
- 이 순간의 변화를 놓고 **순간 변화율**이라는 이름을 붙임
- **순간변화율**은 어느쪽을 향하는 방향성을 지니고 있으므로, 이 방향을 따라 직선을 그려주면 그래프와 맞닿는 접선이 그려지는데, 이선이 점에서의 **기울기**가 됨

- **미분**을 한다는 것은 쉽게 말해 이 **‘순간 변화율’**을 구한다는 것임
- **미분 계수**: 어느 순간에 어떤 변화가 일어나고 있는지 를 숫자로 나타낸 것
- 미분계수는 그래프에서의 **기울기**를 의미함


- 여기서 Δ(델타)는 변화량을 나타내는 기호

- 이 그래프에서 x값의 증가량은 b-a, y값의 증가량은 f(b) - f(a)

- 이때 직선 AB의 기울기를 A와 B사이의 **'평균 변화율'**이라 함
- **순간변화율**: x 증가량 △x이 0에 가까울 만큼 아주 작을 때의 순간적인 기울기


- 기울기는 y값의 증가량 / x값의 증가량 순간기울기는  이것을 로 쓸 수 있음
- 함수 f(x)를 미분하라는 것을 f'(x)또는 f(x)로 표기함


### 딥러닝을 공부하는 과정 중에 자주 만나게 되는 중요한 4가지 미분
- f(x) = a 에서 a가 상수일때 미분 값 0
- f(x) = x 일때 미분 값 1
- f(x) = ax에서 a가 상수이면 미분 값은 a
- f(x) = x^a에서 a가 자연수이면 미분 값은 ax^(a-1)이 됨

**편미분**
- 여러 가지 변수가 식 안에 있을 때, 모든 변수를 미분하는 것이 아니라 우리가 원하는 변수만 미분 그 외에는 모두 상수로 취급

- f(x, y) = x^2 + yx + a 에 관해 편미분 하는 과정은 미분의 성질 4번에 따라 x^2는 2x가됨
- x에 관해 미분하면 다른 모든 항은 상수로 취금하므로 y는 상수




**지수**
- a를 ‘밑’이라 하고 ☐를 ‘지수’
- **지수함수** : 변수 x가 지수 자리에 있는 경우

- 지수 함수에서는 밑 a의 값이 중요
- a > 1 or 0 < a < 1


- **시그모이드 함수** : 지수 함수에서 밑의 값이 자연 상수 e인 함수를 말함



**로그**

- 로그 함수와 지수 함수는 서로 역함수
