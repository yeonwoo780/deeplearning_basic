## 퍼셉트론
- **인공신경망** : 뉴런과 뉴런의 연결처럼 퍼셉트론의 연결을 통해 입력값에 대한 판단을 하게 하는 것<br><br>

- **신경망의 기본구조**는여러 층의 퍼셉트론을 서로 연결시키고 복잡하게 조합하여 주어진 입력 값에 대한 판단을 하게 하는 것<br><br>

- **퍼셉트론**은 입력 값과 활성화 함수를 사용해 출력 값을 다음으로 넘기는 가장 작은 **신경망 기본 단위**<br><br>



- 기울기 a나 y절편 b와 같은 용어를 퍼셉트론 개념에 맞춰 딥러닝 처럼 표현하면<br><br>
 - $ y = ax + b $ (a는 기울기, b는 y절편) 가
 - $ y = wx + b $ (w는 가중치, b는 바이오스)<br><br>

- 가중합(weighted sum) : 입력 값 x 와 가중치 w의 곱을 모두 더한 다음 거기에 바이오스 b 더한 값<br><br>

- 활성화 함수(activation function) : 가중합의 결과를 놓고 1 또는 0을 출력하여 다음으로 보내는데, 여기서 0과 1을 판단하는 함수 (sigmoid, relu, softmax등)

```python
from sklearn.linear_model import Perceptron
# 샘플과 레이블이다. 
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0, 0, 0, 1]
# 퍼셉트론을 생성한다. tol는 종료 조건이다. random_state는 난수의시드이다.
clf = Perceptron(tol=1e-3, random_state=0)
# 학습을 수행한다. 
clf.fit(X, y)
# 테스트를 수행한다. 
print(clf.predict(X))
```
