## 오차 역전파

- 신경망 내부의 가중치는 오차 역전파 방법을 사용해 수정<br><br>

- 오차역전파는 경사하강법의 확장개념<br><br>

- 가중치를 구하는 방법은 경사 하강법을 그대로 이용하면 됨<br><br>

- **오차역전파** : 임의의 가중치를 선언하고 결과값을 이용해 오차를 구한뒤 이 오차가 최소인 지점으로 계속해서 조금씩 이동시킴(최적화 과정)<br><br>

- 이 오차가 최소가 되는 점(미분시 기울기 0)을 찾으면 그것이 바로 정답<br><br>

- **오차 역전파(back propagation)** : 다층 퍼셉트론에서의 최적화 과정<br><br>

- **오차 역전파 구동방식**은 다음과 같이 정리할 수 있음<br><br>
 1. 임의의 초기 가중치 w를 준뒤 결과 $y_out$을 계산<br><br>
 2. 계산 결과와 우리가 원하는 값 사이의 오차 구함<br><br>
 3. 경사 하강법을 이용해 바로 앞 가중치를 오차가 작아지는 방향으로 가중치 업데이트<br><br>
 4. 위 과정을 더이상 오차가 줄어들지 않을 때 까지 반복<br><br>
