## 모델 설계하기

```python
# 딥러닝을 구동하는 데 필요한 케라스 함수 불러옴
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

Data_set = np.loadtxt("../dataset/ThoraricSurgery.csv", delimiter=",")
print(Data_set)
print(Data_set.shape)

X = Data_set[:, 0:17] # 환자의 기록 17개의 입력 값(독립변수)
Y = Data_set[:, 17] # 수술 결과

model = Sequential()
model.add(Dense(30, input_dim=X.shape[1], activation = 'relu'))# 17개의 값을 받아 은닉층의 30개 노드로 보낸다
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, Y, epochs = 100, batch_size = 10)
# epoch : 학습 프로세스가 모든 샘플에 대해 한 번 실행되는 것
# 100 epoch이므로 각 샘플이 처음부터 끝까지 100번 재사용될 때까지 학습
# batch_size=10은 전체 470개 샘플을 10개씩 끊어서 집어넣으라는 뜻
```
